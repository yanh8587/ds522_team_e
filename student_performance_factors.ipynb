```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('StudentPerformanceFactors.csv')
df.head()
```

```python
for col in df:
    # Check if the data type of the column is object type
    if df[col].dtype == 'O':
        sns.countplot(x=col,data=df,palette='Set1')  # Set1 is a common colorful palette
        plt.show()
```

```python
# Create a new figure with specified size
plt.figure(figsize=(8,8))

# Generate a scatter plot with a linear regression line
sns.lmplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.show()
```

```python
# Create a violin plot to show the distribution of exam scores for different family income levels
sns.violinplot(x='Family_Income', y='Exam_Score', data=df, palette='Set1')
plt.title('Family Income vs Exam Score')
plt.show()
```

```python
sns.violinplot(x='Peer_Influence', y='Exam_Score', data=df, palette='Set1')
plt.title('Peer Influence vs Exam Score')
plt.show()
```
